```python
def barchart(feature,xlabel):
    plt.figure(figsize=[14,8]) # Creates a new figure with a specified size
    by_feature =  df1.groupby([feature], as_index=False)['Classes'].sum() # Groups the dataframe by the feature and calculates the sum of 'Classes'
    ax = sns.barplot(x=feature, y="Classes", data=by_feature[[feature,'Classes']], estimator=sum, palette="pastel")# Creates a bar plot
    ax.set(xlabel=xlabel, ylabel='Fire Count') # Sets the labels for the x and y axes
```

```python
plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot = True)
```

```python
def correlation(dataset , threshold):
  col_corr = set()
  corr_matrix= dataset.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[i,j])>threshold:
        colname = corr_matrix.columns[i]
        col_corr.add(colname)
  return col_corr
```

```python
corr_features = correlation(X_train,0.85)
```

```python
corr_features
```

```python
X_train.drop(corr_features, axis=1, inplace=True)
X_test.drop(corr_features, axis=1, inplace=True)
X_train.shape, X_test.shape
```

## Feature Scaling Or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

## Box Plots To understand Effect Of Standard Scaler

```python
plt.subplots(figsize = (15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('X_train Before Standard Scaling')
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Standard Scaling')
```

```python
from sklearn .linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
linreg = LinearRegression()
linreg.fit(X_train_scaled,y_train)
y_pred = linreg.predict(X_test_scaled)
```

```python
mea = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Mean_Absolute_Error", mea)
print("R2 Score", score)
plt.scatter(y_test,y_pred)
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso = Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred = lasso.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean Abs Error",mae)
print("R2 Score", score)
plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(X_train_scaled,y_train)
```

```python
y_pred = lassocv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
```

```python
mae_lassocv = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Mean Absolute Error", mae_lassocv)
print("R2 Score", score)
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred=ridge.predict(X_test_scaled)
mea = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Mean Absolute Error", mea)
print("R2 Score", score)
plt.scatter(y_test,y_pred)
```
